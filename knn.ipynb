```python
#seperation of xa nd y 
x=df.iloc[:,:-1]
y=df['price_range']
```

```python
x
```

```python
y
```

```python
# splitiing into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=123)
```

```python
# model building 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypreed=knn.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypreed)
```

```python
param_grid={"n_neighbors":[3,5,7,9],
            "metric":['euclidean','manhattan','minkowski'],
            "weights":['uniform','distance']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gd=GridSearchCV(knn,param_grid,scoring="accuracy")
```

```python
gd.fit(xtrain,ytrain)
```

```python
gd.best_params_
```

```python
gd.best_score_
```

```python
knn1=KNeighborsClassifier(metric='euclidean', n_neighbors= 9, weights= 'distance')
knn1.fit(xtrain,ytrain)
ypred=knn1.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypreed)
```
